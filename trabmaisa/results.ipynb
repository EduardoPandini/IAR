```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
df = pd.read_csv('Airbnb_Data(1).csv')

print(df.head())
```

```python
df = df.drop_duplicates()
print(df.head())
colunas_para_remover = ['NAME', 'host name', 'country', 'house_rules', 'license']
df = df.drop(columns=colunas_para_remover)

print(df.head())
```

```python

colunas_verificadas = ['country code']
df = df.dropna(subset=colunas_verificadas)
print(df.head())
```

```python

colunas_a_preencher = ['host_identity_verified']
df[colunas_a_preencher] = df[colunas_a_preencher].fillna('unconfirmed')
print(df.head())
```

```python
colunas_a_normalizar = ['reviews per month', 'calculated host listings count']
for coluna in colunas_a_normalizar:
    df[coluna] = (df[coluna] - df[coluna].min()) / (df[coluna].max() - df[coluna].min())
print(df.head())
```

```python
mapeamento = {'flexible': 1, 'moderate': 0.5, 'strict': 0}
coluna_a_modificar = 'cancellation_policy'
df[coluna_a_modificar] = df[coluna_a_modificar].replace(mapeamento)
print(df.head())
```

```python


coluna_especifica = 'instant_bookable'

encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(df[[coluna_especifica]])

one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out([coluna_especifica]))

dados_processados = pd.concat([df, one_hot_df], axis=1)

dados_processados.head()


df = dados_processados
df = df.drop(columns=coluna_especifica)
```

```python
df.to_csv('dados_processados.csv', index=False)
```
